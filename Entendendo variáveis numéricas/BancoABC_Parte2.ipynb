```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
dados = pd.read_csv('/content/dados.csv')
del dados ['Unnamed: 0']
dados
```

```python
descriçao = dados['Score'].describe()
descriçao = pd.DataFrame(descriçao)
descriçao.index = ['Contagem', 'Média', 'Desvio Padrão', 
                   'Mínimo', '25%', 'Mediana', '75%',
                   'Máximo']
descriçao
```

```python
sns.displot(dados, x = dados['Score'], height = 7, aspect = 2,
            kind = 'hist', legend = True)
```

```python
sns.boxplot(x = dados['Score'])
```

```python
score_min = dados.loc[dados['Score']<400]
score_min = pd.DataFrame(score_min)
score_min
```
